```python
#Librerías para este módulo
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Carga del dataset y visualización
df = pd.read_csv('C://Users//USER//Desktop//Data Science//edX//DA Python - UNI VALENCIA//Módulo 2//Datos_y_archivos_M2//CovidCVMes21.csv')
df
```

```python
#Para tamaño: linewidth, markersize
plt.plot(df.Mes, df.Casos, linestyle = '--', marker='o', linewidth=1, markersize=5, label="Casos", color= 'r')
#Para la cuadricula de fondo usar GRID
plt.grid(linestyle="-.", linewidth=0.5, color='0.65', zorder=-10)
plt.title('Casos Totales Covid')
plt.xticks(rotation=45)
plt.tick_params(axis='x', length=6, pad=5)
plt.xlabel('Mes')
plt.ylabel('Casos')
plt.legend()
plt.show()
```

### Ejes agrupados en una sola gráfica

```python
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.grid(linestyle="-.", linewidth=0.5, color='0.65', zorder=-10)
ax1.set_ylabel("Hospitalizados", fontsize=14, color='b')
ax1.plot(df.Mes, df.Hosp, marker='o', linestyle='dashed', color='blue', linewidth=1.2, markersize=5, label="Hosp")
ax1.set_xticklabels(df.Mes, rotation = 45)
ax1.set_ylim(0,14000)

#Con esto agrupamos los ejes
ax2 = ax1.twinx()

ax2.set_ylabel("UCI x10", fontsize=14, color='r')
ax2.plot(df.Mes, df.UCI, marker='>', linestyle='dashed', color='red', linewidth=1.2, markersize=5, label="UCI")
ax2.set_ylim(0,1400)


#Para que las leyendas no se sobrepongan debemos agregar la 2da al 1ro:
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2)
```

- fig, ax = plt.subplot(filas,columnas)

El orden/estructura que considero correcta es:
- set_ylabel: nombre del gráfico-label, para mostrar el nombre en el eje (la leyenda se hace con el label del grafico sea plot, bar, etc)
- grid para el fondo (cuadricula)
- set_xticklabels(df.Mes, rotation = 45): para la serie que se usara como nombre de las MARCAS en el EJE X
- Límites de las marcas si tuvieran: set_ylim()
- tipo de grafico: plot, bar, etc (label=''): Importante el label para usar legend

```python
#Gráfico usando Figura y SUBPLOTS (Distintos ejes dentro de la figura)
fig, ax = plt.subplots(2,2, figsize = (20,18))

ax[0,0].set_ylabel("Casos", fontsize=14, color='b')
ax[0,0].grid(linestyle="-.", linewidth=0.5, color='.25')
ax[0,0].set_xticklabels(df.Mes, rotation = 45)
ax[0,0].set_ylim(0,df.Casos.max()*1.1)
ax[0,0].plot(df.Mes, df.Casos, marker='o',  linestyle='--',  linewidth=1.5, markersize=8, label='Casos')
ax[0,0].legend()
```

### Para no copiar el código 4 veces, podemos simplificarlo y hacer una función

```python
#A= fila, b= columna, nombre=nombre del df['nombre'], col=color
def crear_grafico(a, b, nombre, col):
    ax[a,b].set_title(nombre + " Covid-19 mensuales CV 2021", fontsize=20, verticalalignment='bottom')
    ax[a,b].set_ylabel(nombre, fontsize=18, color=col)
    ax[a,b].set_xlabel('Meses', fontsize=18, color=col)
    ax[a,b].grid(linestyle="-.", linewidth=0.5, color='.25')
    ax[a,b].set_xticklabels(df.Mes, rotation = 45)
    ax[a,b].set_ylim(0,df[nombre].max()*1.1)
    ax[a,b].plot(df.Mes, df[nombre], marker='o',  linestyle='--',  linewidth=1.5, markersize=8, label='Casos', color=col)
    ax[a,b].legend()
```

```python
fig, ax = plt.subplots(2,2, figsize = (28,20))
plt.suptitle("Datos Covid-19 CV 2021", fontsize=24)
crear_grafico(0,0, 'Casos', 'r')
crear_grafico(0,1, 'Hosp', 'b')
crear_grafico(1,0, 'UCI', 'g')
crear_grafico(1,1, 'Fallecidos', 'c')
plt.subplots_adjust(wspace=0.2, hspace=0.4)
plt.show()
```
